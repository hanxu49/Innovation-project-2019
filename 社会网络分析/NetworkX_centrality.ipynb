```
import itertools
import copy
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt

# Grab edge list data hosted on Gist
edgelist = pd.read_csv('relation.csv')
# Preview edgelist
edgelist.head(10)

G = nx.from_pandas_edgelist(edgelist, 'n1', 'n2', edge_attr ='frequency') 
  

# Grab node list data hosted on Gist
nodelist = pd.read_csv('node.csv')
# Preview nodelist
nodelist.head(5)
```

```
import pandas as pd
dc = pd.DataFrame.from_dict(nx.degree_centrality(G), orient="index")
bc = pd.DataFrame.from_dict(nx.betweenness_centrality(G), orient="index")
cc = pd.DataFrame.from_dict(nx.closeness_centrality(G), orient="index")
dc.to_csv("degree_centrality.csv")
bc.to_csv("betweenness_centrality.csv")
cc.to_csv("closeness_centrality.csv")
```
